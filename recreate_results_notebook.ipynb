```python
def geom_prog(min=1, max=100, step=20, integer=False):
    a = (max/min)**(1/(step-1))
    if integer:
        l = [round(min*a**i) for i in range(step)]
    else:
        l = [min*a**i for i in range(step)]
    return l
```

```python
listt = geom_prog(min=0.1, max=2, step=20, integer=False)
```
