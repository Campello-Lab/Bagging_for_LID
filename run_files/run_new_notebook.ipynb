```python
import math
import numpy as np


def geometric_progression(a: float, b: float, n: int):
    """Return n values of a geometric progression from a to b (inclusive)."""
    if n < 2:
        return [a] if n == 1 else []
    r = (b / a) ** (1 / (n - 1))          # common ratio
    return [a * r**k for k in range(n)]

gp = geometric_progression(0.05, 1.0, 7)
gp[-1] = 1
print(-np.sort(-np.array(gp)))
print(np.round(np.sort(1/np.array(gp))))
```

```python
def geometric_progression2(a, c, r, n):
    def iterate(a, c, I):
        res = a
        if I > 0:
            for i in range(I):
                res = res*r + c
        return res
    return [iterate(a, c, k) for k in range(n)]
```
